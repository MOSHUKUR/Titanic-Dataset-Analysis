### Data Cleaning*

```python
df.columns.values
```

```python
# Some cloumns aren't necessary for our stats.
#We better delete them
unwanted_columns= ['Cabin','Ticket','Name','PassengerId']
df.drop(unwanted_columns, axis=1, inplace=True)
df.shape
```

```python
# Drop the 2 records where embarked value is Null
df = df[pd.notnull(df['Embarked'])]
```

```python
#Time to check for Nulls
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
#To check if there is a movie with a non-logical release year
df.Embarked.unique()
```

```python
df.Survived.unique()
```

```python
df.Sex.unique()
```

```python
#Change the embarked keys to better readable ones
df.loc[:,'Embarked'].replace(['C','S','Q'], ['Cherbourg','Southampton','Queenstown'], inplace=True)
#And the survived keys
df.loc[:,'Survived'].replace([0,1],['No','Yes'],inplace=True)
df.loc[:,'Pclass'].replace([1,2,3],['First','Second', 'Third'],inplace=True)
```

#####  S = Southampton, C = Cherbourg, and Q = Queenstown

## Exploratory Data Analysis

```python
#Make a dataframe for non missing 'Age'values
age_clean = df[(df['Age'].notnull())]
age_clean['Age'] = age_clean['Age'].astype(int)
```

```python
sns.displot(df['Age'].dropna(),bins=30, kde=True);
```

```python
age_stats = age_clean['Age'].describe().astype('int64')
age_stats
```

```python
# Get summary descriptive statistics
age_clean['age_cat'] = pd.cut( x=age_clean["Age"], bins=[0, age_stats['25%'], age_stats['75%'], np.inf],labels=["Child", "Youth", "Adult"] )
age_clean.head(3)
```

```python
sns.countplot(x= age_clean.age_cat,color='blue', alpha=0.4)
sns.countplot(x= age_clean['age_cat'],  hue = df['Survived'])
plt.xlabel("Categorial Ages");
```

# Class

```python
df.Pclass.value_counts().plot.pie(autopct='%.2f%%').legend(labels=["3rd Class","1st Class","2nd Class"], loc='center right', bbox_to_anchor=(2.25, 0.5)).set_title("Passenger classes")
```

```python
sns.countplot(x = 'Embarked', data = df, color='blue', alpha=0.4).legend(df.Embarked.value_counts().values)
sns.countplot(x= df['Embarked'],  hue = df['Pclass'])
plt.legend(title= "Passenger Class")
plt.ylabel("Passsengers Count")
plt.xlabel("Embark Port")
plt.show()
```

```python
fare_stats = df.Fare.describe()
fare_stats
```

```python
df['fare_cat'] = pd.cut( x=df["Fare"], bins=[0, fare_stats['25%'], fare_stats['50%'],fare_stats['75%'], np.inf],labels=["Broke", "Poor", "Rich", "Wealthy"] )
df.head(4)
```

```python
sns.countplot(x = "fare_cat", hue = "Survived", data = df).legend(labels = ["Deceased", "Survived"])
plt.ylabel("Count")
plt.xlabel("Passengers Social Class")
plt.show()
```

```python
df_2 = df.copy()
df_2.loc[:,'Survived'].replace(['No','Yes'],[0,1],inplace=True)
sns.lmplot(data = df_2, x='Age', y='Survived', hue = 'Sex', aspect = 2)
plt.show()
```

### Surprisingly grown up women had better chace to survive unlike men of same age!

```python
survival_by_class = sns.catplot(data = df_2, x='Pclass',y='Survived', kind = 'bar', aspect = 2)
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()
```

### As expected people who rode the first class had better chance of survival
